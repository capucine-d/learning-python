# Tests 

###  if

if est suivi d'une bloc d'instruction, lui même défini par une indentation.

if condition:
- Instruction A

else:
- Instruction B

```python
x = 3
if x < 10: # La condition est une expression booléenne: elle peut prendre pour valeur vrai ou faux
    print(x)
else: print(x)
```

## Boucles

### For
Boucle bornée: quand on sait combien de fois doit avoir lieu la répétition, on utilise généralement une boucle for.

for condition:
- instruction A

```python
for i in [0, 1, 2, 3]:
    print("i a pour valeur", i)
```

### While
Boucle non bornée: Si on ne connait pas à l’avance le nombre de répétitions, on choisit une boucle while.

while Condition:
- Instruction A

```python
x = 1
while x < 10:
    print("x a pour valeur", x)
    x = x * 2
print("Fin")
```

### Break
L’instruction break permet de « casser » l’exécution d’une boucle (while ou for). Elle fait sortir de la boucle et passer à l’instruction suivante.

```python
a = 2
while a <= 10:
    print("ok")
    break
```

### Continue
L’instruction continue permet de passer prématurément au tour de boucle suivant. Elle fait continuer sur la prochaine itération de la boucle.

```python
for num in range(2, 10):
    if num % 2 == 0:
        print ("Found an even number", num)
        continue
    print ("Found a number", num)
```

### Else
La clause else dans un boucle permet de définir un bloc d’instructions qui sera exécuté à la fin seulement si la boucle s’est déroulée complétement sans être interrompue par un break.

```python
for n in range(2, 8):
    for x in range(2, n):
        if n % x == 0:
            print(n, "egale", x, "*", n/x)
            break
    else:
        print(n, "est un nombre premier")
```

# Fonctions 

def nom_fonction(liste de paramètres):
- bloc d'instructions

```python
def compteur_complet(start, stop, step):
    i = start
    while i < stop:
        print(i)
        i = i + step

compteur_complet(1, 7, 2)
```

#### Global
Cette instruction permet d’indiquer - à l’intérieur de la définition d’une fonction - quelles sont les variables à traiter globalement.

```python
def test():
    global b
    b = 5
    print(a, b)
    
a = 2
b = 7
test()
print(a, b)
```
